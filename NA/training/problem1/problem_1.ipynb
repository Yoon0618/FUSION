# 1. 지구 대기의 질량은 얼마인가? - 수치적분

악독하신 교수님이 불쌍한 대학원생에게 "이 자료로 지구 대기의 질량을 구해오라"고 하셨습니다. 주어진 자료는 지구의 반지름($R_{earth}=6378$ km)과 대기의 밀도 분포뿐입니다. 

<div style="text-align: center">
  <img src="img/Atmosphere_model.png" width="600">
  <p><strong>Figure 1.</strong> 고도에 따른 대기의 밀도와 온도. x-축이 고도이고, y-축이 밀도와 온도이다. 이런 자료로 고도가 주어지면 밀도를 구할 수 있다.</p>
</div>

대학원생은 대기의 질량을 구하는 방법을 고민하기 시작했습니다. 처음에는 대기의 부피와 밀도를 곱하면 될 것이라 생각했지만, 대기의 밀도가 고도에 따라 달라져 올바른 계산이 아니란 것을 깨달았죠. 다시 고민에 빠진 대학원생은 다음 두가지 아이디어를 떠올렸습니다.

<br>

<strong>첫번째</strong>, "대기를 작은 부피들의 모음으로 생각하자". 대기의 밀도는 고도($r$)에 따라 달라지지만, 충분히 작은 부피(예를 들어 주먹 크기)의 대기를 생각하면, 그 부피 $dV$ 안의 밀도는 거의 일정할 것입니다. 따라서 그 부피의 질량 $dm$은 다음과 같이 구할 수 있습니다.

$$
dm(\vec{r}) = \rho (\vec{r}) dV(\vec{r})
$$

여기서 $\vec{r}$은 대기의 위치 벡터입니다. 이 아이디어를 바탕으로, 대학원생은 대기의 질량을 구하기 위해 모든 미소 부피의 질량을 더하면 된다는 것을 깨달았습니다.

$$
M = \int dm(\vec{r}) = \int \rho (\vec{r}) dV(\vec{r}) \tag{1}
$$

<br>

<strong>두번째</strong>, "같은 고도에 존재하는 대기의 부피들은 밀도가 동일하므로, 고도에 따라 대기를 층으로 나누자". $dm(r)$을 고도가 $r$에 존재하는 모든 대기의 질량이라고 하면, 아래와 같이 쓸 수 있습니다.

$$
dm(r) = \rho (r) dV(r)
$$

같은 고도라면 $\vec{r}$의 크기는 동일하므로 벡터 $\vec{r}$ 대신 고도 $r$를 사용해 $dV$를 쓸 수 있습니다. $dV$의 모습은 아래 그림과 같겠죠.

<div style="text-align: center">
  <img src="img/spherical shell.svg" width="600">
  <p><strong>Figure 2.</strong> dv(r)은 반지름 r과 R 사이의 부피이다. r과 R이 충분히 비슷해야 첫번째 아이디어가 성립한다.</p>
</div>

이제 $dV(r)$을 구해보죠. $dV(r)$은 매우 얇은 구의 껍질이 $dr$만큼의 두께를 가지는 부피입니다. 따라서 $dV(r)$은 구의 겉넓이 $4 \pi r^2$에 두께 $dr$을 곱한 것과 같습니다. 즉,

$$
dV(r) = 4 \pi r^2 dr \tag{2}
$$

입니다. 이를 (1)에 대입하면, 대기의 질량은 다음과 같이 쓸 수 있습니다.

$$
M = \int _{r_0}^{r_1} dm(r) = \int _{r_0}^{r_1} \rho (r) dV(r) = \int _{r_0}^{r_1} \rho (r) 4 \pi r^2 dr \tag{3}
$$

정적분이므로 적분 구간을 대기가 존재할 구간으로 설정해줘야 하겠죠. 따라서 지표면 $r_0$, 그보다 높은 고도(아마 우주) $r_1$으로 적어주었습니다. 이제 (3)을 계산하면 대기의 질량을 구할 수 있습니다. 하지만, (3)은 적분이므로, 적분을 계산할 방법이 필요합니다. $\rho(r)$은 부정적분을 구할 수 있는 함수가 아닙니다. 따라서, 대학원생은 수치적분을 사용하기로 합니다.

먼저 정적분의 정의를 보면,

$$
\int_a^b f(x) dx = \lim_{n \to \infty} \sum_{i=1}^{n} f(x_i) \Delta x
$$

와 같습니다. 여기서 $x_i$는 $[a, b]$ 구간을 $n$개의 작은 구간으로 나누었을 때, $i$번째 구간 안의 값입니다. $\Delta x$는 각 구간의 길이입니다. 실제로 $n$을 무한대로 할 수는 없으므로, 충분히 큰 $n$을 사용해서 근사적인 값을 구합니다. 따라서, (3)을 수치적으로 계산하기 위해서는 $r$을 적당히 나누고, 각 구간의 오른쪽 끝점에서의 밀도를 계산한 후, 이를 모두 더하면 됩니다.

````python
import numpy as np # 숫자 계산에 유용한 도구를 제공하는 패키지
import pymsis # 대기 밀도 데이터를 제공하는 패키지
from datetime import datetime # 시간 관련 패키지

# 먼저 계산에 필요한 상수들을 정의한다.
R_earth = 6378.0 # km
pi = 3.141592 # 원주율
r_start = R_earth # 적분은 지표면부터 시작
r_end = R_earth + 1000 # 고도 1000 km까지 적분
n = 10000 # 적분 구간을 n 등분한다.

# 고도(km)를 입력하면 대기 밀도(km/m^3)를 출력하는 함수를 정의한다. 코드를 이해할 필요는 없다.
# 예를 들어, density(10)의 값은 4.129e-01 kg/m³이다.
# 시간과 대기의 경도, 위도도 입력할 수 있지만 특정 시간, 위치로 고정한다.
def density(alt):
    # 고정 시간: 2025-01-01
    date = np.datetime64(datetime(2025, 1, 1))
    # 고정 위치: 적도 (0°, 0°)
    lat, lon = 0.0, 0.0
    # pymsis 패키지의 calculate 함수로부터 대기 밀도 값을 얻는다.
    data = pymsis.calculate(date, lon, lat, alt, version='2.1')
    
    rho = data[0][0] # calculate 함수는 대기 밀도 뿐만 아니라 다른 잡다한 데이터도 반환하므로 리스트 인덱싱으로 밀도만 rho에 저장한다.
    return float(rho)

rs = np.linspace(r_start, r_end, n) # R_start부터 R_end까지를 n 등분해서 r_i들의 리스트를 만든다.

M = 0 # 대기의 총 질량 M을 우선 0으로 두고, dm을 하나씩 더해나는 방식으로 적분을 계산한다.
for r in rs:
    dr = (r_end - r_start )/n
    dV = 4*pi*r**2*dr

    # density는 고도를 입력으로 받는다. r은 고도(지표면으로부터의 거리)가 아닌 지구 중심으로부터의 거리이므로, r이 아닌 r-R_earth를 입력한다.
    rho = density(r-R_earth)*10**9 # density의 값의 단위는 kg/m^3이므로 10^9을 곱해서 kg/km^3으로 변환
    dm = rho*dV
    M = M + dm # 미소 질량을 총 질량에 더한다.
    print(f"계산 진행률 : {r - r_start:.2f} km / {r_end - r_start:.2f} km, 현재 질량: {M:.2e} kg")  # 진행 상황 출력

print(f"지구 대기의 질량은 약 {M} kg입니다.")
````

## 대기 밀도는 고도 뿐만 아니라 위도와 경도에 따라 달라진다
대학원생은 대기의 밀도가 고도에 따라 달라진다고 생각했지만, 사실 대기의 밀도는 위도와 경도에 따라서도 달라집니다. 따라서, 대기의 밀도를 구할 때는 위도와 경도를 고려해야 합니다. 물론 그 차이는 미미하므로 대략적으로만 구하려 한다면 무시해도 됩니다. 하지만 조금 더 엄밀하게 가보자고요. 

대학원생의 두번째 아이디어는 잘못되었으니, 고도뿐만 아니라 위도($\theta$)와 경도($\phi$)를 고려해서 대기의 밀도 $dm$을 구해야 합니다.

$$
\rho(\vec{r}) = \rho(r, \theta, \phi) \\
dm(\vec{r}) = \rho (r, \theta, \phi) dV(r, \theta, \phi)
$$

<div style="text-align: center">
  <img src="img/spherical_coordinate.png" width="600">
  <p><strong>Figure 4.</strong> 고도와 경도, 위도는 구면 좌표계의 r, φ, θ에 대응한다.</p>
</div>

학부 미적분학2에서 배우는 내용을 참조하면, $dV(r, \theta, \phi)$는 다음과 같이 쓸 수 있습니다(증명 생략).

$$
dV(r, \theta, \phi) = r^2 \sin(\theta) dr d\theta d\phi \tag{4}
$$

<div style="text-align: center">
  <img src="img/volume_in_spherical.png" width="600">
  <p><strong>Figure 5.</strong> 구면좌표계에서 미소 부피</p>
</div>

전체 질량 $M$을 다시 써봅시다.

$$
M = \int dm(\vec{r}) = \int \rho (r, \theta, \phi) dV(r, \theta, \phi) = \iiint \rho (r, \theta, \phi) r^2 \sin\theta dr d\theta d\phi \tag{5}
$$

식이 조금 복잡해졌지만, 단순히 적분을 3번 하면 됩니다. 즉, $r$, $\theta$, $\phi$에 대해 각각 적분을 하면 됩니다.

$$
M = \int_{r_0}^{r_1} \int_{\theta_0}^{\theta_1} \int_{\phi_0}^{\phi_1} \rho (r, \theta, \phi) r^2 \sin\theta dr d\theta d\phi
$$

더 쉽게 쓰면, 

$$
M = \int_{r_0}^{r_1} \left( \int_{\theta_0}^{\theta_1} \left( \int_{\phi_0}^{\phi_1}
\rho(r, \theta, \phi)\, r^2 \sin\theta \, d\phi \right) d\theta \right) dr
$$

대기는 지구 사방에 존재하므로 $\phi_0 = 0$, $\phi = 2\pi$, $\theta_0 = 0$, $\theta_1 = \pi$로 설정할 수 있습니다. $r$은 지구의 반지름 $R_{earth}$에서 시작해서, 대기가 존재하는 최대 고도(대략 1000 km)까지 적분합니다.

- 보조자료 1: 다중적분의 파이썬 구현

<br>
식 (5)와 같이 여러번 적분하는 것을 다중적분, 연속적분이라 합니다. 다중적분과 연속적분에 익숙하지 않으신 분들을 위해 이중적분 파이썬 구현 코드를 보여드리겠습니다. 적분이 다음과 같다고 해봅시다.

$$
I = \int_a^b \int_c^d f(x, y) \, dy \, dx
$$

```python
import numpy as np

a, b = 0, 10
c, d = 0, 5
xs = np.linspace(a, b, 100)  # x에 대한 분할
ys = np.linspace(c, d, 100)  # y에 대한 분할

delta_x = (b-a)/len(xs)  # x의 간격
delta_y = (d-c)/len(ys)  # y의 간격
I = 0  # 적분 결과를 저장할 변수

def f(x, y):
    # f(x, y)를 정의합니다.
    return x * y  # 예시 함수로 x와 y의 곱을 사용합니다.

for x in xs:
    for y in ys:
        # f(x, y) 계산
        f_value = f(x, y)
        f_dx_dy = f_value * delta_x * delta_y  # f(x, y) * dA 계산
        # I에 f(x, y) * dA를 더함
        I += f_dx_dy
print(I)
```

````python
import numpy as np # 숫자 계산에 유용한 도구를 제공하는 패키지
import pymsis # 대기 밀도 데이터를 제공하는 패키지
from datetime import datetime # 시간 관련 패키지

# 먼저 계산에 필요한 상수들을 정의한다.
R_earth = 6378.0 # km
pi = 3.141592 # 원주율
r_start = R_earth # 적분은 지표면부터 시작
r_end = R_earth + 1000 # 고도 1000 km까지 적분
n_r = 10000 # r 적분 구간을 n 등분한다.
n_theta = 60 # theta 적분 구간을 1도씩, 180 등분한다.
n_phi = 120 # phi 적분 구간을 360등분한다.

# 고도(km)를 입력하면 대기 밀도(kg/m^3)를 출력하는 함수를 정의한다. 코드를 이해할 필요는 없다.
# 예를 들어, density(10)의 값은 4.129e-01 kg/m³이다.
def density(alt, lat, lon): # 고도 alt_km, 위도 lat, 경도 lon을 입력받아 대기 밀도를 반환
    # 고정 시간: 2025-01-01
    date = np.datetime64(datetime(2025, 1, 1))
    # 구하고자 하는 대기의 경도 위도를 도(deg) 단위로 받는다. theta와 phi는 호도법(radian) 단위이므로 단위 변환이 필요하다.
    lat = np.rad2deg(lat)  # 위도를 라디안에서 도로 변환
    lon = np.rad2deg(lon)  # 경도를 라디안에서 도로 변환
    # pymsis 패키지의 calculate 함수로부터 대기 밀도 값을 얻는다.
    data = pymsis.calculate(date, lon, lat, alt, version='2.1')
    rho = data[0, 0]
    return float(rho)

rs = np.linspace(r_start, r_end, n_r) # R_start부터 R_end까지를 n 등분해서 r_i들의 리스트를 만든다.
phis = np.linspace(0, 2*pi, n_phi)
thetas = np.linspace(0, pi, n_theta)

dr = (r_end - r_start )/len(rs)  # r 적분 구간의 길이를 n_r로 나눈다.
dphi = 2*pi/len(phis)  # phi 적분 구간의 길이를 n_phi로 나눈다.
dtheta = pi/len(thetas)  # theta 적분 구간의 길이를 n_theta로 나눈다.

M = 0 # 대기의 총 질량 M을 우선 0으로 두고, dm을 하나씩 더해나는 방식으로 적분을 계산한다.
for r in rs:
    for theta in thetas:
        for phi in phis:
            dV = r**2 * dr * np.sin(theta) * dphi * dtheta

            lat = pi/2 - theta # 위도는 적도에서 0이지만, theta는 북극에서 시작하므로 변환 필요
            lon = phi
            # density는 고도를 입력으로 받는다. r은 고도(지표면으로부터의 거리)가 아닌 지구 중심으로부터의 거리이므로, r이 아닌 r-R_earth를 입력한다.
            rho = density(r-R_earth, lat, lon)*10**9 # density의 값의 단위는 kg/m^3이므로 10^9을 곱해서 kg/km^3으로 변환
            dm = rho*dV
            M = M + dm # 미소 질량을 총 질량에 더한다.
        # print(f"현재 r: {r} km, theta: {theta}, M: {M} kg")  # 진행 상황 출력
    print(f"계산 진행률 : {r - r_start:.2f} km / {r_end - r_start:.2f} km, 현재 질량: {M:.2e} kg")  # 진행 상황 출력

print(f"지구 대기의 질량은 약 {M} kg입니다.")
````

# 1. 공기 저항을 포함하는 포물체 운동

## 1.1. 공기 저항을 무시한 포물체 운동

일반물리학1에서 배우셨겠지만 공기 저항을 무시하면 포물체는 포물선(2차 곡선)의 궤적을 그리며 날라갑니다. 이건 $\vec{F}=m\vec{a}$로부터 쉽게 유도되죠. 한번 유도 해보겠습니다.

위 그림과 같이 물체가 $\vec{v_0}$의 속도로 발사되는 경우를 생각해봅시다. 작용하는 힘은 공기 저항을 무시했으니 중력 $m\vec{g}$가 유일합니다. 따라서, 운동 방정식($\vec{F}=m\vec{a}$)은 다음과 같겠죠.

$$
\begin{align*}
m\vec{a}=\vec{F}
\end{align*}
$$

가속도 $\vec{a}$에 대해서 정리해보면,

$$ 
\begin{align*}
m\vec{a}=m\vec{g} \\
\vec{a}=\vec{g} \tag{1}
\end{align*}
$$

가 됩니다. 중력가속도는 상수이니 (우리가 잘 아는 대로) 이 운동은 등가속도 운동임을 알 수 있죠. 어떤 물체의 운동을 안다는 것은 시간에 따른 위치를 아는 것이죠. 가속도 $\vec{a}$는 위치 $\vec{x}$을 시간으로 두번 미분한 것으로 정의되니, (1)번을 두번 시간으로 적분하면 위치가 구해질 겁니다. 먼저 한번 적분하고 다시 적분해볼게요.

$$
\int_{t_0}^{t} \vec{a} dt = \int_{t_0}^{t} \vec{g} dt \tag{2}
$$

적분을 하려면 적분 구간이 필요합니다. (2)번의 적분 구간은 $t_0$에서 $t$까지입니다. $t_0$는 초기 시간, 즉 발사 순간이니 편의상 $t_0=0$으로 놓겠습니다. t는 우리가 위치를 알고 싶은 특정 시간입니다. 계산을 해보면,

$$
\vec{v}(t) -  \vec{v}(0) = \int_{0}^{t} \vec{g} dt = \vec{g} \int_{0}^{t}  dt = \vec{g}t \tag{3}
$$

이 됩니다. 여기서 $\vec{v}(0)$는 초기 속도 $\vec{v_0}$이니, (3)번을 정리하면 다음과 같습니다.

$$
\vec{v}(t) = \vec{v_0} + \vec{g}t \tag{4}
$$

다시 한번 적분해서 위치를 구해보겠습니다. 위 과정과 동일하게 (4)번을 시간에 대해 적분하면 위치를 얻을 수 있습니다.

$$
\int_{0}^{t} \vec{v}(t) dt = \int_{0}^{t} \vec{v_0} dt + \int_{0}^{t} \vec{g} t dt = \vec{v_0} \int_{0}^{t} dt + \vec{g} \int_{0}^{t} t dt \\
\vec{x}(t) - \vec{x}(0) = \vec{v_0} t + \vec{g} \frac{t^2}{2} \\
\vec{x}(t) = \vec{x_0} + \vec{v_0} t + \vec{g} \frac{t^2}{2} \tag{5}
$$

역시 $\vec{x_0}=\vec{x}(0)$임을 이용했습니다. 이제 임의의 시간 $t$에 대한 위치 $\vec{x}(t)$를 구했습니다. 이 식을 이용하면 포물체 운동의 궤적을 그릴 수 있겠죠.

## 1.2. 공기 저항을 포함한 포물체 운동

이제 공기 저항을 받는 포물체 운동을 생각해봅시다. 보통 느린 속도에서는 공기 저항은 속력에 비례하고(선형 공기 저항이라고 부름), 이동 방향에 반대 방향으로 작용합니다. 한편 야구공처럼 빠르게 운동하는 물체에는 속력의 제곱에 비례하는 저항력(제곱 공기 저항이라고 부름)이 작용합니다. 방향은 역시 반대 방향이고요. 따라서, 공기 저항을 포함한 운동 방정식은 다음과 같이 쓸 수 있습니다.

$$
\begin{align*}
m\vec{a} = m\vec{g} - b\vec{v} \; (느린 속도의 경우) \\
m\vec{a} = m\vec{g} - c|v|\vec{v} \; (빠른 속도의 경우)
\end{align*}
$$

여기서 b, c는 속도에 대한 저항력의 비례 상수입니다. 선형 공기 저항의 경우 앞선 계산과 비슷한 과정으로 운동 방정식을 풀 수 있습니다. 하지만 제곱 공기 저항의 경우 이것이 불가능합니다. 다른 말로하면 단순히 시간에 대해서 적분해서는 위치를 구할 수 없습니다.

$$
\begin{align*}
\int_{0}^{t} \vec{a} dt = \int_{0}^{t} \left( \vec{g} - \frac{b}{m}\vec{v} \right) dt \\
\vec{v}(t) - \vec{v_0} = \vec{g}t - \frac{c}{m} ( \vec{x}(t) - \vec{x_0}) \\
\int_{0}^{t} \vec{a} dt = \int_{0}^{t} \left( \vec{g} - \frac{c}{m}|v|\vec{v} \right) dt
\end{align*}
$$
